# 🟢 EXEMPLO 1: Básico

## Chat Simples com MangabaAgent

O `MangabaAgent` é a forma mais simples de usar o Mangaba. Ele já vem com os protocolos **A2A** (Agent-to-Agent) e **MCP** (Model Context Protocol) habilitados automaticamente.

```python
# Criar agente (usa configurações do .env automaticamente)
from mangaba_ai import MangabaAgent

agent = MangabaAgent()

print(f"🤖 Agente: {agent.agent_id}")
print(f"   MCP: {agent.mcp_enabled} | A2A: {agent.use_a2a}")
```

### 💬 Chat com Contexto Automático

O protocolo MCP permite que o agente **lembre** das conversas anteriores:

```python
# Primeira mensagem - apresentação
resposta1 = agent.chat("Olá! Meu nome é Maria e sou cientista de dados.")
print("💬 Resposta 1:")
print(resposta1)
```

```python
# Segunda mensagem - testando a memória do agente (protocolo MCP em ação!)
resposta2 = agent.chat("Qual é o meu nome e qual é a minha profissão?")
print("💬 Resposta 2 (testando memória):")
print(resposta2)
```

### 🔍 Análise de Texto

O agente pode analisar textos com instruções personalizadas:

```python
# Texto para análise
texto = """
O mercado de tecnologia continua em expansão no Brasil.
Startups de inteligência artificial receberam investimentos recordes este ano.
O setor de e-commerce cresceu 30% comparado ao ano anterior.
Novas regulamentações estão sendo discutidas para o setor digital.
"""

# Analisar o texto
analise = agent.analyze_text(texto, "Extraia os 3 pontos principais deste texto")

print("🔍 Análise do Texto:")
print(analise)
```

### 🌐 Tradução

```python
# Tradução de texto
texto_pt = "A inteligência artificial está transformando o mundo dos negócios."

traducao = agent.translate(texto_pt, target_language="en")
print(f"Original (PT): {texto_pt}")
print(f"Tradução (EN): {traducao}")
```

### 📋 Resumo de Contexto

```python
# Resumindo texto
contexto_longo = """
O projeto Mangaba AI é um framework de orquestração multi-agente.
Ele suporta múltiplos provedores de LLM como Google Gemini, OpenAI e Anthropic.
Os principais componentes são: Agent, Task, Crew e Process.
O protocolo A2A permite comunicação entre agentes.
O protocolo MCP gerencia memória e contexto.
"""

resumo = agent.summarize(contexto_longo)
print(f"Resumo: {resumo}")
```

## 🚀 EXEMPLO 2: Crew Multi-Agente (Complexo)

Neste exemplo avançado, vamos criar um **Crew** (equipe) de múltiplos agentes especializados que trabalham juntos para realizar uma tarefa complexa.

### Cenário: Pesquisa e Relatório sobre IA

Vamos criar uma equipe com:
1. **Pesquisador**: Coleta informações sobre o tema
2. **Analista**: Analisa e sintetiza as informações
3. **Escritor**: Produz o relatório final

### 📦 Importações para Crew

```python
from mangaba.core import Agent, Task, Crew, Process

# Verificar importações
print("✅ Componentes importados:")
print(f"   - Agent: {Agent}")
print(f"   - Task: {Task}")
print(f"   - Crew: {Crew}")
print(f"   - Process: {Process}")
```

### 👥 Criando os Agentes Especializados

```python
# 🔍 Agente Pesquisador
pesquisador = Agent(
    role="Pesquisador de IA",
    goal="Coletar informações relevantes e atualizadas sobre inteligência artificial",
    backstory="""Você é um pesquisador experiente com vasta experiência em 
    tecnologia e inteligência artificial. Sua especialidade é encontrar 
    informações precisas e relevantes sobre qualquer tema tecnológico.""",
    verbose=True,
    allow_delegation=False
)

print(f"✅ Pesquisador criado: {pesquisador.role}")
```

```python
# 📊 Agente Analista
analista = Agent(
    role="Analista de Dados",
    goal="Analisar e sintetizar informações para extrair insights valiosos",
    backstory="""Você é um analista de dados sênior com forte capacidade 
    analítica. Você transforma dados brutos em insights acionáveis e 
    identifica padrões e tendências importantes.""",
    verbose=True,
    allow_delegation=False
)

print(f"✅ Analista criado: {analista.role}")
```

```python
# ✍️ Agente Escritor
escritor = Agent(
    role="Escritor Técnico",
    goal="Produzir relatórios claros, bem estruturados e envolventes",
    backstory="""Você é um escritor técnico premiado com habilidade de 
    transformar informações complexas em textos acessíveis e interessantes.
    Seu estilo é claro, conciso e profissional.""",
    verbose=True,
    allow_delegation=False
)

print(f"✅ Escritor criado: {escritor.role}")
```

### 📝 Definindo as Tarefas

```python
# Tarefa 1: Pesquisa
tarefa_pesquisa = Task(
    description="""Pesquise sobre os principais avanços em IA generativa em 2024.
    Foque em:
    - Novos modelos de linguagem (LLMs)
    - Aplicações práticas em empresas
    - Tendências emergentes
    
    Forneça uma lista com pelo menos 5 pontos principais.""",
    expected_output="Uma lista detalhada dos principais avanços em IA generativa",
    agent=pesquisador
)

print(f"✅ Tarefa de pesquisa criada")
```

```python
# Tarefa 2: Análise (depende da pesquisa)
tarefa_analise = Task(
    description="""Com base na pesquisa realizada, analise:
    - Quais avanços têm maior impacto no mercado?
    - Quais são os riscos e oportunidades?
    - Como as empresas podem se preparar?
    
    Forneça uma análise estruturada com insights acionáveis.""",
    expected_output="Análise detalhada com insights e recomendações",
    agent=analista,
    context=[tarefa_pesquisa]  # Depende da tarefa de pesquisa
)

print(f"✅ Tarefa de análise criada (depende da pesquisa)")
```

```python
# Tarefa 3: Relatório (depende da análise)
tarefa_relatorio = Task(
    description="""Produza um relatório executivo completo incluindo:
    - Resumo executivo (3-4 parágrafos)
    - Principais descobertas
    - Recomendações estratégicas
    - Conclusão
    
    O relatório deve ser profissional e adequado para apresentação a executivos.""",
    expected_output="Relatório executivo completo e bem formatado",
    agent=escritor,
    context=[tarefa_pesquisa, tarefa_analise]  # Depende de ambas as tarefas anteriores
)

print(f"✅ Tarefa de relatório criada (depende de pesquisa e análise)")
```

### 🚀 Criando e Executando o Crew

```python
# Criar o Crew com processo sequencial
crew = Crew(
    agents=[pesquisador, analista, escritor],
    tasks=[tarefa_pesquisa, tarefa_analise, tarefa_relatorio],
    process=Process.SEQUENTIAL,  # Executa tarefas em sequência
    verbose=True
)

print("✅ Crew criado com sucesso!")
print(f"   - Agentes: {len(crew.agents)}")
print(f"   - Tarefas: {len(crew.tasks)}")
print(f"   - Processo: {crew.process}")
```

```python
# 🎬 Executar o Crew (kickoff)
# NOTA: Isso requer uma API key válida configurada
# Descomente a linha abaixo para executar:

# resultado = crew.kickoff()
# print(resultado)

# Para demonstração, vamos mostrar a estrutura do crew
print("📋 Estrutura do Crew:")
print("-" * 50)
for i, (agent, task) in enumerate(zip(crew.agents, crew.tasks), 1):
    print(f"\n{i}. Agente: {agent.role}")
    print(f"   Objetivo: {agent.goal[:50]}...")
    print(f"   Tarefa: {task.description[:50]}...")
```

## 🔗 Protocolos: A2A e MCP

O Mangaba AI suporta dois protocolos importantes:

### 📡 A2A (Agent-to-Agent)
Permite comunicação direta entre agentes para colaboração.

### 🧠 MCP (Model Context Protocol)
Gerencia memória e contexto das interações.

```python
# Exemplo de uso dos protocolos
from protocols.a2a import A2AProtocol
from protocols.mcp import MCPProtocol

# Criar protocolo A2A
a2a = A2AProtocol(agent_id="coordenador")
print(f"✅ A2A Protocol criado: {a2a.agent_id}")

# Criar protocolo MCP
mcp = MCPProtocol()
print(f"✅ MCP Protocol criado")
```

```python
# Conectar agentes via A2A (usando o agente criado anteriormente)
# O MangabaAgent já tem A2A integrado, então usamos ele diretamente
agent.a2a_protocol.connect_agent(agent)  # Conecta consigo mesmo para demonstração

print(f"🔗 Agentes conectados: {list(agent.a2a_protocol.connected_agents.keys())}")
```

```python
# Armazenar contexto no MCP
from protocols.mcp import MCPContext, ContextType, ContextPriority

# Criar sessão
session_id = mcp.create_session("sessao_tutorial")

# Criar e adicionar contexto
contexto_obj = MCPContext.create(
    context_type=ContextType.CONVERSATION,
    content={
        "tema": "IA Generativa",
        "agentes": ["pesquisador", "analista", "escritor"],
        "status": "completo"
    },
    priority=ContextPriority.HIGH,
    tags=["tutorial"]
)
context_id = mcp.add_context(contexto_obj, session_id)

# Recuperar contexto
contexto = mcp.get_context(context_id)
print(f"📦 Contexto armazenado: {contexto.content}")
```
